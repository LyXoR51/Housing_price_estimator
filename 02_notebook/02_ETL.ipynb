# 1. Libraries
---

```python
from sqlalchemy import create_engine, Table, Column, Integer, MetaData, Float, TEXT, Date, BIGINT, INT
from datetime import datetime
from dotenv import load_dotenv
import pandas as pd
import os

load_dotenv()
```

# 2. Extract
---

```python
df = pd.read_csv('../data/real_estate_dataset.csv')
df.head()
```

# 3. Transform
---

```python
# date
now = datetime.now()
ts = int(datetime.timestamp(now)* 1000)
df['created_at'] = ts

df.head()
```

```python
# lower columns title
df.columns = df.columns.str.lower()
df.head()
```

```python
# index 
df = df.set_index('id')
df.head()
```

# 4. Load
---

```python
# Start engine
postgres_database = os.environ["HOUSING_PRICES_DATABASE"]
engine = create_engine(postgres_database, echo=True)

# create table
metadata = MetaData()
table_name = 'housing_prices'

my_table = Table(
    table_name, metadata,
    Column('id', Integer, primary_key=True),
    Column('square_feet', Float),
    Column('num_bedrooms', Integer),
    Column('num_bathrooms', Integer),
    Column('num_floors', Integer),
    Column('year_built', Integer),
    Column('has_garden', Integer),
    Column('has_pool', Integer),
    Column('garage_size', Integer),
    Column('location_score', Float),
    Column('distance_to_center', Float),
    Column('price', Float),
    Column('price_predict', Float),
    Column('created_at', TEXT)
)

#run the engine
metadata.create_all(engine)
```

```python
# load data
df.to_sql(table_name, engine, if_exists="append")
```
